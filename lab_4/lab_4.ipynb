### Use pointwise mutual information to compute the measure for all pairs of words.


````python
total_word_pairs = df_bigrams_count['count'].sum()
````

````python
total_word_pairs
````

````python
total_words = df_unigrams_count['count'].sum()
````

````python
total_words 
````

````python
def count_pmi(bigram):
    count = bigram_counts[(bigram)]
    words = bigram.split(' ')
    if len(words) != 2:
        return 0
    word1, word2 = words[0], words[1]
    p_x_y = count / total_word_pairs
    p_x = unigram_counts[(word1)] / total_words
    p_y = unigram_counts[(word2)] / total_words
    if p_x == 0 or p_y == 0:
        return 0
    return math.log2(p_x_y / (p_x * p_y))
````

````python
df_bigrams_count['bigram_pmi'] = df_bigrams_count['ngram'].apply(count_pmi)
````

### 7. Use SpaCy to [lemmatize](https://spacy.io/api/lemmatizer) and [tag](https://spacy.io/api/tagger) the sentences in the corpus.

````python
def lemmatize_and_tag(sentence):
    doc = nlp(sentence)
    lemmas = [token.lemma_ for token in doc if token.is_alpha]
    tags = [token.pos_ for token in doc if token.is_alpha]
    return lemmas, tags
````

````python
# df = pd.read_csv('processed_data.csv')
df
````

````python
df['lower'] = df['lower'].astype(str)
df[['lemmas', 'tags']] = df['lower'].apply(lemmatize_and_tag).apply(pd.Series)
````

````python
import ast

df['lemmas'] = df['lemmas'].apply(lambda lemmas: [lemma.strip("''") for lemma in lemmas])
df['tags'] = df['tags'].apply(lambda tags: [tag.strip("''") for tag in tags])
````

### 8. Using the tagged corpus compute bigram statistic for the tokens containing:
####   a. lemmatized, downcased word
####  b. morphosyntactic category of the word (subst, fin, adj, etc.)

### 9. For example: "Ala ma kota.", which is tagged as:
   ```
   Ala	Ala	subst:sg:nom:f
   ma	   mieć	fin:sg:ter:imperf
   kota	kot	subst:sg:acc:m2
   .     .	interp
   ```
   #### the algorithm should return the following bigrams: `ala:subst mieć:fin` and `mieć:fin kot:subst`.

````python
def compute_tagged_bigrams(n, lemmas, tags):
    ngrams = []
    for idx in range(len(lemmas)-n+1):
        ngram_lemmas = tuple(lemmas[idx:idx+n])
        ngram_tags = tuple(tags[idx:idx+n])
        ngram = tuple(zip(ngram_lemmas, ngram_tags))

        ngrams.append(ngram)

    return ngrams
````

````python
lematized_ngrams = []
tags_ngrams = []

for lemmas, tags in zip(df['lemmas'], df['tags']):
    lemmatized_tags = list(zip(lemmas, tags))
    lematized_ngrams.extend(compute_tagged_bigrams(2, lemmas, tags))
    tags_ngrams.extend(tags)
````

### 11. Group the bigrams by morphosyntactic tag, i.e. a pair of words belongs to a given group if all pairs have the same syntactic category for the first and the second word. E.g. one group would be words with `subst` as the first words and `adj` as the second word.


````python
df_lematized_bigram_counts_5['pos_tags'] = df_lematized_bigram_counts_5['bigram'].apply(lambda x: tuple(tag[1] for tag in x))
df_lematized_bigram_counts_5
````

````python
grouped_bigrams = df_lematized_bigram_counts_5.groupby('pos_tags')['bigram_count'].sum().reset_index()
grouped_bigrams = grouped_bigrams.rename(columns={'bigram_count': 'bigrams_group_count'})
grouped_bigrams['bigrams_in_group'] = df_lematized_bigram_counts_5.groupby('pos_tags').apply(lambda x: list(x[['bigram', 'bigram_count']].itertuples(index=False, name=None))).reset_index(name='bigrams_in_group')['bigrams_in_group']
````
